```python
def scoring(result, choice):
    global total_points
    if result == "win":
        total_points += 6
    elif result == "tie":
        total_points += 3
    if choice == "rock":
        total_points += 1
    elif choice == "paper":
        total_points += 2
    elif choice == "scissors":
        total_points += 3
```

```python
def rock(me):
    if me == "X": #lose
        scoring("lose","scissors")
    elif me == "Y": #tie
        scoring("tie", "rock")
    elif me == "Z":#win
        scoring("win", "paper") 
```

```python
def paper(me):
    if me == "X":
        scoring("lose", "rock")
    elif me == "Y":
        scoring("tie", "paper")
    elif me == "Z":
        scoring("win", "scissors")
```

```python
def scissors(me):
    if me == "X":
        scoring("lose", "paper")
    elif me == "Y":
        scoring("tie", "scissors")
    elif me == "Z":
        scoring("win", "rock") 
```

```python
total_points = 0 
with open('input.txt') as f:
    plays=f.readlines()
    data=[tuple(play.strip().split()) for play in plays]
    for elf, me in data:
        if elf == "A": #rock
            rock(me)
        elif elf == "B": #paper
            paper(me)
        elif elf == "C": #scissors
            scissors(me)
    print("Total points: ",total_points)
```

```python
X - means I need to lose 
Y - means I need to draw
Z - means I need to win 
```
